```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scale =1.0
exponential_data = np.random.exponential(scale, size=100000)
```

```python
plt.hist(exponential_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000) * 5, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red", label="Normal")
```

```python
plt.axis([0, 10, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
```

```python
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Z = X^2 + Y^2')
```
