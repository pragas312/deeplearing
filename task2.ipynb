```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
df = pd .read_csv("Au_nanoparticle_dataset.csv")
```

```python
new_df =df[['N_total','N_bulk','N_surface','R_avg']]
```

```python
print(new_df.head(20))
```

```python
stats =new_df.describe().transpose()[['mean','std','25%','50%','75%']]
print(stats)
```

```python
fig,axes =plt.subplots(1,4,figsize=(20,5))
for i, column in enumerate(new_df.columns):
    sns.histplot(new_df[column],ax = axes[i],kde=True)
    axes[i].set_title(column)
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
```
